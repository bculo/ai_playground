```python
"""
    Import packages to use model
"""

import os
from transformers import pipeline
from PIL import Image
```

```python
os.listdir("./images")
```

```python
"""
    Show image
"""
Image.open(r"./images/image_one.jpg") 
```

```python
images = [
    Image.open(r"./images/image_one.jpg"),
    Image.open(r"./images/image_three.jpg")
]
```

```python
"""
    Model microsoft/resnet-50
    
    Input
    random_image.jpg
    
    Result
    {'score': 0.598035991191864, 'label': 'valley, vale'}
    {'score': 0.33323296904563904, 'label': 'alp'}
    {'score': 0.03781522437930107, 'label': 'lakeside, lakeshore'}
    {'score': 0.010467279702425003, 'label': 'promontory, headland, head, foreland'}
    {'score': 0.0023355211596935987, 'label': 'cliff, drop, drop-off'}
"""

pipe = pipeline("image-classification", model="microsoft/resnet-50")

for image in images:
    classification_result = pipe(image)
    for result in classification_result:
        print(result)
    print("\n")
```
